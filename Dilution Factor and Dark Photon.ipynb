```python
m_h = 125.3 #GeV
v = 246 #GeV
Gammah = 4.6e-3 #GeV
def GammahTo2DP(m_V,kp):#Output in GeV
    return kp**2/(64*mt.pi)*v**2/m_h*np.sqrt(1-4*(m_V/m_h)**2)*(1 - 4*(m_V/m_h)**2 + 12*(m_V/m_h)**4)
def BRhTo2DP(m_V,kp):
    return GammahTo2DP(m_V,kp)/Gammah
def kpFunc(BR,m_V):
    return np.sqrt(Gammah*BR*(64*mt.pi)/v**2*m_h/np.sqrt(1-4*(m_V/m_h)**2)/(1 - 4*(m_V/m_h)**2 + 12*(m_V/m_h)**4))
BRhTo2DPplot, BRhTo2DPAx = plt.subplots(figsize = (15,7))
N = int(1e5)
M_V = [0.5, 20, 30, 40]
BR = [0.1, 0.01, 0.001]
KpTABLE = np.zeros((3,len(M_V)))
i = 0
for BRp in BR:
    j = 0
    for m_V in M_V:
        KpTABLE[i][j] = kpFunc(BRp,m_V)
        j +=1
    i += 1
print(KpTABLE)
kp = np.linspace(1e-6,1e-1,N)
color = 'darkviolet'
for m_V in M_V:
    BRhTo2DPAx.plot(kp, BRhTo2DP(m_V,kp), color, label= r'$m_{A^\prime}$ = '+str(m_V)+' GeV')
BRhTo2DPAx.hlines(0.1, 1e-5, 1e-1, ls = '--', color = 'k', label = r'$10\%$')
BRhTo2DPAx.hlines(0.01, 1e-5, 1e-1, ls = '--', color = 'k', label = r'$1\%$')
BRhTo2DPAx.hlines(0.001, 1e-5, 1e-1, ls = '--', color = 'k', label = r'$0.1\%$')
# BRhTo2DPAx.vlines()
# plt.text(1e-3, 6.5e-4, r'$m_{A^\prime} = 2\cdot 10^{-2}$', color = color, fontsize = 15)
# plt.text(1e-3, 6.5e-6, r'$m_{A^\prime} = 10^{-2}$', color = color, fontsize = 15)
# plt.text(1e-3, 6.5e-8, r'$m_{A^\prime} = 10^{-1}$', color = color, fontsize = 15)
# plt.text(1e-3, 6.5e-10, r'$m_{A^\prime} = 10$', color = color, fontsize = 15)
# plt.text(1e-3, 6.5e-12, r'$m_{A^\prime} = 50$', color = color, fontsize = 15)
BRhTo2DPAx.set_xscale('log')
BRhTo2DPAx.set_yscale('log')
BRhTo2DPAx.tick_params(axis='both', which = 'major', length = 10.0, labelsize = 19)
BRhTo2DPAx.tick_params(axis='both', which = 'minor', length = 5.0)
BRmax = 0.2
BRmin = 0.0005
BRhTo2DPAx.set_ylim(BRmin,BRmax)
BRhTo2DPAx.set_xlim(9e-4,3e-2)
xvals = [2e-3, 3.5e-3, 7.5e-3, 1.3e-2, 1e-2]
BRhTo2DPAx.plot(kp, BRhTo2DP(10,kp), 'green', label= r'$m_{A^\prime}$ =10 GeV')
labelLines(BRhTo2DPAx.get_lines(), xvals = xvals, fontsize = 11)
BRhTo2DPAx.set_title('BR of the SM Higgs into two dark photons', fontsize = 20)
BRhTo2DPAx.set_xlabel(r'$\kappa^\prime$', fontsize = 20)
BRhTo2DPAx.set_ylabel(r'$BR \left(h \rightarrow A^\prime A^\prime\right)$', fontsize = 20)
BRhTo2DPAx.grid()
# m10GeV = plt.Line2D([], [], color='green', markersize = 25, label=r'$m_{A^\prime}$ = 10 GeV')
# BRhTo2DPAx.legend(handles = [m10GeV], fontsize = 12.5, loc = 'upper left')
plt.savefig('SM Higgs BR into two dark photons.png')
plt.show()
```
